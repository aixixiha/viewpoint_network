```python
import os
import numpy as np
import matplotlib.pyplot as plt

# 数据路径
p = "/home/sduu39/zHongchang/data/label/6_Computational Linguistics/pairwise_subset_similarities_all28328.npy"
vals = np.load(p)
```

```python
# plt.figure(figsize=(7, 4.5), dpi=150)
color = (0/255, 102/255, 204/255)

plt.hist(vals, bins=200, range=(0,1),
         color=color, alpha=0.75,
         edgecolor="white", linewidth=0.3)

mean_val = vals.mean()
plt.axvline(mean_val, color="red", linestyle="--", linewidth=1.2)

plt.title("Computational Linguistics", fontsize=14, weight="bold")
plt.xlabel("Cosine Similarity", fontsize=14); plt.ylabel("Frequency", fontsize=14)
plt.grid(alpha=0.3, linestyle="--")
plt.tight_layout()
plt.savefig("sim/6_Computational Linguistics.png", dpi=600)
plt.show()
```
