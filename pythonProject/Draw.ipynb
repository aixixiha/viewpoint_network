```python
import os
import numpy as np
import matplotlib.pyplot as plt

# 数据路径
p = "/home/sduu39/zHongchang/data/label/6_Computational Linguistics/pairwise_subset_similarities_all28328.npy"
vals = np.load(p)
```

```python
plt.figure(figsize=(7, 4.5), dpi=150)
color = (0/255, 102/255, 204/255)

plt.hist(vals, bins=200, range=(0,1),
         color=color, alpha=0.75,
         edgecolor="white", linewidth=0.3)

mean_val = vals.mean()
plt.axvline(mean_val, color="red", linestyle="--", linewidth=1.2)

plt.title("Similarity Distribution", fontsize=14, weight="bold")
plt.xlabel("Cosine Similarity"); plt.ylabel("Frequency")
plt.grid(alpha=0.3, linestyle="--")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.colors as mcolors

plt.figure(figsize=(7, 4.5), dpi=150)

for alpha in np.linspace(0.3, 0.9, 5):
    plt.hist(vals, bins=200, range=(0,1),
             color=(30/255,144/255,255/255),
             alpha=alpha*0.4)

plt.title("Gradient Histogram", fontsize=14, weight="bold")
plt.xlabel("Cosine Similarity"); plt.ylabel("Frequency")
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import gaussian_kde

plt.figure(figsize=(7,4.5), dpi=150)

# 直方图
plt.hist(vals, bins=200, density=True, range=(0,1),
         color=(135/255,206/255,250/255),
         edgecolor="white", alpha=0.6)

# KDE 曲线
kde = gaussian_kde(vals)
xs = np.linspace(0,1,500)
plt.plot(xs, kde(xs), color='blue', linewidth=2)

plt.title("Histogram + KDE", fontsize=14, weight="bold")
plt.xlabel("Cosine Similarity"); plt.ylabel("Density")
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```
