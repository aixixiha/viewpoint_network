```python
import json
import argparse
import re
```

```python
pattern = re.compile(
    r"\[Sentence \d+\]\s*([\s\S]*?)\s*\[Extracted Viewpoints in Sentence \d+\]\s*\[(.*?)\]",
    re.DOTALL
)
```

```python
text = "[Sentence 1]\nThis chapter addresses the challenge of applying relational database technologies to manage taxonomies, which are commonly used to classify documents, knowledge and websites into a hierarchy of topics.\n[Extracted Viewpoints in Sentence 1]\n[This chapter addresses the challenge of using relational database technologies to manage taxonomies that are commonly used to classify documents, knowledge, and websites into a hierarchy of topics.]\n\n[Sentence 2]\nIt first describes how denormalizing the data model can facilitate data retrieval from such topic hierarchies.\n[Extracted Viewpoints in Sentence 2]\n[Denormalizing the data model can facilitate data retrieval from taxonomies used to classify documents, knowledge, and websites into a hierarchy of topics.]\n\n[Sentence 3]\nIt then shows how the typical data maintenance difficulties associated with denormalized data models can be solved using database triggers.\n[Extracted Viewpoints in Sentence 3]\n[The typical data maintenance difficulties associated with denormalized data models can be solved using database triggers.]\n[Using database triggers, the data maintenance difficulties associated with denormalized data models can be addressed.]\n\n[Sentence 4]\n[No additional sentences were provided.]"
```

```python
total_papers = 0
total_viewpoints = 0
not_matched = 0
with open(input_file, 'r', encoding='utf-8') as fin, open(output_file, 'w', encoding='utf-8') as fout:
    for line in fin:
        line = line.strip()
        if not line:
            continue
        try:
            obj = json.loads(line)
        except Exception as e:
            print("JSON 解析错误，跳过：", e)
            print(line)
            print("------------------------------------------")
            continue

        view_text = obj.get("viewpoints", None)
        if not view_text:
            continue

        matches = pattern.findall(view_text)
        if not matches:
            not_matched += 1
            print(line)
        total_papers += 1
        for sentence, viewpoint in matches:

            viewpoint = viewpoint.strip()
            sentence = sentence.strip()
            if not sentence or not viewpoint:
                continue

            new_obj = {
                "viewpoint": viewpoint,
                "sentence": sentence
            }
            for key, value in obj.items():
                if key != "viewpoints" :
                    new_obj[key] = value

            fout.write(json.dumps(new_obj, ensure_ascii=False) + "\n")
            total_viewpoints += 1
        if total_papers % 10000 == 0:
            print(total_papers, total_viewpoints)
    print(f"共处理论文 {total_papers} 篇，提取 viewpoint {total_viewpoints} 条。")
```
