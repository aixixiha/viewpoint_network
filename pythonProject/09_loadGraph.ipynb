# 查看图的属性特征

```python
import pickle

g_path = "/home/sduu39/zHongchang/data/networkx/view_similarity_graph_tau060.gpickle"   # 你的实际路径

with open(str(g_path), "rb") as f:        # 注意是 "rb" 二进制读
    G_tau = pickle.load(f)

print(type(G_tau))
```

```python
import networkx as nx
if nx.is_connected(G_tau):
    Gcc = G_tau
else:
    largest_cc_nodes = max(nx.connected_components(G_tau), key=len)
    Gcc = G_tau.subgraph(largest_cc_nodes).copy()

print("最大连通子图：节点数", Gcc.number_of_nodes(), "边数", Gcc.number_of_edges())
```

```python
density = nx.density(G_tau)
print("Density:", density)
```

```python
avg_clustering = nx.average_clustering(G_tau)
print("Average clustering coefficient:", avg_clustering)
```

```python
topk = sorted(G_tau.degree, key=lambda x: x[1], reverse=True)[:10]
```

```python
for node, deg in topk:
    print(node, deg)
```

```python
for node, deg in topk:
    attrs = G_tau.nodes[node]
    print(attrs["viewpoint"])
```

```python
avg_path_length = nx.average_shortest_path_length(Gcc)
print("平均最短路径长度 (LCC):", avg_path_length)
```

```python
import pickle

g_path = "/home/sduu39/zHongchang/data/networkx/view_similarity_graph_tau065.gpickle"  # 你的实际路径

with open(str(g_path), "rb") as f:  # 注意是 "rb" 二进制读
    G_tau = pickle.load(f)

print(type(G_tau))

import networkx as nx

if nx.is_connected(G_tau):
    Gcc = G_tau
else:
    largest_cc_nodes = max(nx.connected_components(G_tau), key=len)
    Gcc = G_tau.subgraph(largest_cc_nodes).copy()

print("最大连通子图：节点数", Gcc.number_of_nodes(), "边数", Gcc.number_of_edges())
density = nx.density(G_tau)
print("Density:", density)

avg_clustering = nx.average_clustering(G_tau)
print("Average clustering coefficient:", avg_clustering)
topk = sorted(G_tau.degree, key=lambda x: x[1], reverse=True)[:10]
for node, deg in topk:
    print(node, deg)
for node, deg in topk:
    attrs = G_tau.nodes[node]
    print(attrs["viewpoint"])
```
