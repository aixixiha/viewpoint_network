```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from tqdm import tqdm
import json
import os

# === 路径配置 ===
pro_path = "/home/sduu39/zHongchang/"
model_path = pro_path + "SentenceTransformers"  # 你的模型目录
data_path = pro_path + "data/"
input_file = data_path + "v4output.json"
output_embedding_file = data_path + "viewpoint_embeddings.npy"

# === 1. 加载模型（完全离线） ===
print("Loading model from local path:", model_path)
model = SentenceTransformer(model_path, device="cuda")
```

```python
input_file = "/home/sduu39/zHongchang/data/v4_107viewExAbstract.jsonl"
output_file = "/home/sduu39/zHongchang/data/v4_embeddings.jsonl"
```

```python
count=0
with open(input_file, 'r', encoding="utf-8") as fin, open(output_file, 'w', encoding="utf-8") as fout:
    for line in fin:
        obj = json.loads(line.strip())
        view = obj.get("viewpoint", None)
        if not view:
            continue
        emb=model.encode(view).tolist()
        obj.update({"embedding": emb})
        fout.write(json.dumps(obj, ensure_ascii=False) + "\n")
        count+=1
        if count%5000==0:
            print(count)
```
